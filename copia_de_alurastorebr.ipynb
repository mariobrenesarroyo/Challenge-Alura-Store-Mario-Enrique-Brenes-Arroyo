### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# prompt: Quiero añadir una nueva columna llamada número de tienda para el url llamado URL el número de tienda será tienda 1, Para el URL dos el número de tiendas será tienda 2 y asi con las demas sucesivamente

loja['Número de tienda'] = 'Tienda 1'
loja2['Número de tienda'] = 'Tienda 2'
loja3['Número de tienda'] = 'Tienda 3'
loja4['Número de tienda'] = 'Tienda 4'
```

```python
#  Como todas las hojas tienen los mismos encabezados me gustaría unirlos en una sola tabla llamada hoja completa

import pandas as pd
hoja_completa = pd.concat([loja, loja2, loja3, loja4])
hoja_completa
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que 'hoja_completa' ya está cargada y tiene columnas como
# 'Número de tienda', 'Preço', y 'Data da Compra'

# Convertir la columna de fecha al tipo datetime si no lo está
hoja_completa['Data da Compra'] = pd.to_datetime(hoja_completa['Data da Compra'])

# --- Gráfico de Ventas Totales por Tienda (Revisado) ---
# Agrupar por 'Número de tienda' y sumar los precios
ventas_por_tienda = hoja_completa.groupby('Número de tienda')['Preço'].sum()

# Obtener los números de tienda (para el eje x del gráfico)
tiendas_ventas = ventas_por_tienda.index

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(tiendas_ventas, ventas_por_tienda, color='skyblue')
plt.xlabel('Número de Tienda')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Tienda (General)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nVentas Totales por Tienda (General):\n", ventas_por_tienda)
```

```python


# Calcular las ventas totales por tienda
ventas_por_tienda = hoja_completa.groupby('Número de tienda')['Preço'].sum()
tiendas_ventas = ventas_por_tienda.index

# Crear el gráfico de barras con etiquetas de datos (redondeadas a 2 decimales)
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas_ventas, ventas_por_tienda, color='skyblue')
plt.xlabel('Número de Tienda')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Tienda (General)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Añadir etiquetas de datos a cada barra (redondeadas a 2 decimales)
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura,
             f'{altura:.2f}',  # Formatear el valor con 2 decimales
             ha='center', va='bottom')

plt.show()

print("\nVentas Totales por Tienda (General):\n", ventas_por_tienda)
```

# 2. Vendas por Categoria


```python

import matplotlib.pyplot as plt
# Calcular la cantidad de ventas por categoría para cada tienda
# Reemplazar 'Categoria' con el nombre correcto de la columna si es diferente
ventas_por_categoria_tienda = hoja_completa.groupby(['Número de tienda', 'Categoria do Produto'])['Preço'].sum().unstack() # Changed 'Categoria' to 'Categoría'

# Crear el gráfico de barras agrupadas horizontalmente
ventas_por_categoria_tienda.plot(kind='barh', figsize=(10, 6))
plt.xlabel('Ventas Totales') # Changed ylabel to xlabel
plt.title('Ventas por Categoría y Tienda')
plt.yticks(rotation=0, ha='right') # Changed xticks to yticks and rotation to 0 for better readability
plt.legend(title='Categoría')
plt.tight_layout()

# Agregar etiquetas de datos a las barras
for i in range(len(ventas_por_categoria_tienda)):
    for j in range(len(ventas_por_categoria_tienda.columns)):
        valor = ventas_por_categoria_tienda.iloc[i, j]
        plt.text(valor, i, f'{valor:.2f}', ha='left', va='center') # Adjusted text positioning for horizontal bars

plt.show()

print("\nVentas por Categoría y Tienda:\n", ventas_por_categoria_tienda)
```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd
import matplotlib.pyplot as plt # Import the matplotlib library


# Calcular el promedio de la evaluación por tienda
promedio_evaluacion_por_tienda = hoja_completa.groupby('Número de tienda')['Avaliação da compra'].mean()

# Obtener los números de las tiendas y los promedios de evaluación
numeros_tiendas = promedio_evaluacion_por_tienda.index
promedios = promedio_evaluacion_por_tienda.values

# Encontrar la tienda con la evaluación promedio más baja y más alta
tienda_min_evaluacion = promedio_evaluacion_por_tienda.idxmin()
min_promedio_evaluacion = promedio_evaluacion_por_tienda.min()
tienda_max_evaluacion = promedio_evaluacion_por_tienda.idxmax()
max_promedio_evaluacion = promedio_evaluacion_por_tienda.max()

print(f"Tienda con la evaluación promedio más baja: {tienda_min_evaluacion} ({min_promedio_evaluacion:.2f})")
print(f"Tienda con la evaluación promedio más alta: {tienda_max_evaluacion} ({max_promedio_evaluacion:.2f})")

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(numeros_tiendas, promedios, color='coral', s=50)
plt.xlabel('Número de Tienda')
plt.ylabel('Promedio de Evaluación de la Compra')
plt.title('Promedio de Evaluación de la Compra por Número de Tienda')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("\nPromedio de Evaluación por Tienda:\n", promedio_evaluacion_por_tienda)
```

# 4. Produtos Mais e Menos Vendidos

```python
# prompt: Hacer un top  N = 10 de productos más vendidos y los menos vendidos con etiquetas de datos

import matplotlib.pyplot as plt
# Obtener los 10 productos más vendidos
top_10_productos = hoja_completa['Produto'].value_counts().head(10)

# Obtener los 10 productos menos vendidos
bottom_10_productos = hoja_completa['Produto'].value_counts().tail(10)

# Crear el gráfico de barras para los productos más vendidos
plt.figure(figsize=(12, 6))
plt.bar(top_10_productos.index, top_10_productos.values, color='skyblue')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.title('Top 10 Productos Más Vendidos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Agregar etiquetas de datos
for i, v in enumerate(top_10_productos.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()


# Crear el gráfico de barras para los productos menos vendidos
plt.figure(figsize=(12, 6))
plt.bar(bottom_10_productos.index, bottom_10_productos.values, color='lightcoral')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.title('10 Productos Menos Vendidos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Agregar etiquetas de datos
for i, v in enumerate(bottom_10_productos.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()

print("\nTop 10 Productos Más Vendidos:\n", top_10_productos)
print("\n10 Productos Menos Vendidos:\n", bottom_10_productos)
```

# 5. Frete Médio por Loja


```python
# prompt: Costo de envío promedio por tienda com etiqueta de datos

import matplotlib.pyplot as plt
# Calculate the average shipping cost per store
average_shipping_cost_per_store = hoja_completa.groupby('Número de tienda')['Frete'].mean()

# Get the store numbers and average shipping costs
store_numbers = average_shipping_cost_per_store.index
averages = average_shipping_cost_per_store.values

# Create the bar chart with data labels
plt.figure(figsize=(10, 6))
bars = plt.bar(store_numbers, averages, color='lightgreen')
plt.xlabel('Numero de tienda')
plt.ylabel('Frete promedio')
plt.title('Costo promedio de envío por tienda:')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Añadiendo leyenda para cada columna bar (rounded to 2 decimal places)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height,
             f'${height:.2f}',
             ha='center', va='bottom')

plt.show()

print("\nCosto promedio de envío por tienda:\n", average_shipping_cost_per_store)
```
